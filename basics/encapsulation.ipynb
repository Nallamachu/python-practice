#### Encapsulation

##### Encapsulation is the concept of wrapping data(variables) and methods(functions) together as a single unit. It restricts direct access to some of the objects components, which is a means of preventing accidental interface and misuse of the data.

```python
class Person:
    def __init__(self, name, age):
        self.name = name    # Public attribute
        self.age = age      # Public attribute

    def __str__(self):
        return f"{self.name} is {self.age} years old."

person1 = Person("Reddy", 30)
print(person1)  # Output: Reddy is 30 years old.
```

```python
dir(person1)
```

##### Defining private variables

##### Below is an example that private variables can not be accessed outside the class

```python
class Person:
    def __init__(self, name, age):
        self.__name = name    # Private attribute
        self.__age = age      # Private attribute

person2 = Person("Reddy", 30)
print(person2.__name)
print(person2.__age)
```

##### Below is an example of accessing private variable through public method

```python
class Person:
    def __init__(self, name, age):
        self.__name = name    # Private attribute
        self.__age = age      # Private attribute

    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age
    
person2 = Person("Reddy", 30)
print(person2.get_name())  # Output: Reddy
print(person2.get_age())   # Output: 30
```

```python
dir(person2)
```

##### Working with protected variables

```python
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

class Employee(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self._employee_id = employee_id

employee1 = Employee("Reddy", 30, "E123")
print(employee1._name)  # Output: Reddy
print(employee1._age)   # Output: 30
print(employee1._employee_id)  # Output: E123
```

##### Below is an example of using getter and setter methods

```python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
    
    def get_name(self):
        return self.__name
    def get_age(self):
        return self.__age
    def set_name(self, name):
        self.__name = name
    def set_age(self, age):
        self.__age = age

person3 = Person("Reddy", 30)
print(person3.get_name())  # Output: Reddy
print(person3.get_age())   # Output: 30
person3.set_name("Subba Reddy")
person3.set_age(31)
print(person3.get_name())  # Output: Subba Reddy
print(person3.get_age())   # Output: 31
```
