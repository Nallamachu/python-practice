#### Lambda functions in python

```python
greet = lambda name: f"Hello, {name}!"
print(greet("Reddy"))

greet_custom = lambda name, greeting: f"{greeting}, {name}!"
print(greet_custom("Reddy", "Hi"))

add_one = lambda x:x+1
print(add_one(5))

sum_all = lambda *args: sum(args)
print(sum_all(1, 2, 3, 4, 5))
```

```python
addition = lambda x, y: x + y
print(addition(10, 20))

substraction = lambda x, y: x - y
print(substraction(10, 5))

multiplication = lambda x, y: x * y
print(multiplication(10, 5))

division = lambda x, y: x / y
print(division(10, 5))
```

##### Below are few examples of using map() and filter() functions with lambda

```python
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

```python
numbers = [1, 2, 3, 4, 5]
evens = list(filter(lambda x: x%2==0, numbers))
print(evens)
```

```python
students = [{"name": "Alice", "grade": 85},
            {"name": "Bob", "grade": 90},
            {"name": "Charlie", "grade": 78},
            {"name": "David", "grade": 92}]
sorted_students = sorted(students, key=lambda student: student["grade"], reverse=True)
print(sorted_students)
```
