#### Conditional Statements in Python

```python
val = int(input())
if(val>0):
    print(f'Given {val} is positive')
elif(val<0):
    print(f'Given {val} is negative')
else:
    print('Given value is zero')
```

```python
# if statement
x = 10
if x > 5:
    print("x is greater than 5")

# if-else statement
y = 3
if y > 5:
    print("y is greater than 5")
else:
    print("y is not greater than 5")

# if-elif-else statement
z = 5
if z > 5:
    print("z is greater than 5")
elif z == 5:
    print("z is equal to 5")
else:
    print("z is less than 5")

# Nested conditionals
age = 25
if age >= 18:
    if age >= 65:
        print("Senior citizen")
    else:
        print("Adult")
else:
    print("Minor")

# Ternary operator
status = "Adult" if age >= 18 else "Minor"
print(status)
```

```python
#### Switch case statement (Python 3.10+)
def switch_case(value):
    match value:
        case 1:
            return "One"
        case 2:
            return "Two"
        case 3:
            return "Three"
        case _:
            return "Other"

print(switch_case(1))  # Output: One
print(switch_case(2))  # Output: Two
print(switch_case(3))  # Output: Three
print(switch_case(4))  # Output: Other  
```
