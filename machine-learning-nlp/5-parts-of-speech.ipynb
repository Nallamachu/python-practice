#### Parts of Speech - POS

```python
paragraph="""
Dr. Kalam was not only a brilliant scientist but also a teacher at heart, dedicating his later years to inspiring students across India.
His vision for a self-reliant, technologically advanced nation was encapsulated in his Vision 2020 plan, which emphasized education, innovation, and sustainable development. 
Known for his humility, he lived a simple life and believed in the power of youth to transform society.
"""
```

```python
import nltk
sentences = nltk.sent_tokenize(paragraph)
print(sentences)
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    print(words)
    tagged = nltk.pos_tag(words)
    print(tagged)
    print()
```

```python
from nltk.corpus import stopwords
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [word for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = " ".join(words) # Converting all the words into sentences again
    tagged = nltk.pos_tag(words)
    print(tagged)
    print()
print(sentences)
```
