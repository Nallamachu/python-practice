#### TF-IDF

- **TF** stands for **Term Frequency**. How often a word appears in a document.
- **IDF** stands for **Inverse Document Frequency**. How unique or rare that word is across all documents.
- TF-IDF together tells you, how important this word in this document compared to other?

##### Core Formulas

- Term Frequency - TF:

```TF(t,d) = \frac{\text{Number of times term } t \text{ appears in document } d}{\text{Total number of terms in document } d}```

- Inverse Document Frequency- IDF:

```IDF(t) = \log \left( \frac{N}{1 + df(t)} \right)```

Where:

N = total number of documents

df(t) = number of documents containing term 𝑡

**Final TF-IDF Score**

```TF-IDF(t,d)=TF(t,d)×IDF(t)```

#### Role of TF-IDF in Generative AI
Even though modern Gen AI models like Transformers rely on embeddings, TF-IDF is still useful in many pipelines.
- Information Retrieval (RAG Systems)
- Text Processing
- Light weight alternative to Embeddings
- Keyword extraction

##### TF-IDF vs Modern Embeddings

| Feature              | TF-IDF      | Embeddings (GenAI) |
| -------------------- | ----------- | ------------------ |
| Type                 | Statistical | Semantic           |
| Understands meaning? | ❌ No        | ✅ Yes              |
| Speed                | ⚡ Fast      | 🐢 Slower          |
| Memory               | Low         | High               |
| Context awareness    | ❌ None      | ✅ Strong           |


##### Limitations
- No understanding of context or symantics 
- Treats words independently
- Fails with
    - Synonyms (car vs vehicle)
    - Polysemy (bank)

````python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "AI is powerful",
    "AI is everywhere",
    "Python is powerful"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(X.toarray())
````
