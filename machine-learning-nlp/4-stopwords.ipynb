#### Stopwatch

```python
paragraph="""
Dr. Kalam was not only a brilliant scientist but also a teacher at heart, dedicating his later years to inspiring students across India.
His vision for a self-reliant, technologically advanced nation was encapsulated in his Vision 2020 plan, which emphasized education, innovation, and sustainable development. 
Known for his humility, he lived a simple life and believed in the power of youth to transform society.
"""
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
stopwords.words('arabic')
```

```python
sentences = nltk.sent_tokenize(paragraph)
print(sentences)
```

```python
type(sentences)
```

```python
stemmer = PorterStemmer()
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = " ".join(words) # Converting all the words into sentences again

print(sentences)
```

#### Using SnowballStemmer

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('english')
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = " ".join(words) # Converting all the words into sentences again

print(sentences)
```

Apply stopwords and filter and then apply snowball stemming

```python
lemmatizer = nltk.WordNetLemmatizer()
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)

print(sentences)
```
