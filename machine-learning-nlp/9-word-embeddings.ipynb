#### Word Embeddings

In Natular Language Processing (NLP), word embedding is a term used for the representation of words for text analysis, typically in the form of a real-valued vector that encodes the meaning of the word such that the words that are closer in the vector space are expected to be similar in meaning.

##### What are word embeddings?
A word embedding is a way of representing words as dense vectors of numbers in a continuous space, where:
- Words with similar meanings have similar vectors
- Semantic relationhips are captured mathematically

##### Why do we need word embeddings?
Traditional methods like BoW or TF-IDF have limitations:
- No understanding of meaning
- ignore word order
- High-dimentional or sparse

Word embeddings solve this by:
- Captiring semantic similary
- Reducing dimentionality
- Enabling better performance in ML/DL models

##### Types of Word Embeddings:
- Count or Frequency
    - One
    - Bag of Words
    - TF-IDF
- Deep Leaning Trained models
    - Word2Vec
        - Continious Bag of Words
        - Skipgram


**As per ChatGPT:**
1. Static Word Embeddings: Each word has a fixed vector regardless of context
    - Word2Vec (By Google):
        - CBOW
        - Skip-gram
    - GloVe (By Stanford University):
        - Uses global word co-occurrences statistics
    - FastText (By Facebook AI Research):
        - Handles Subwords -> better for rare words

2. Contexual Word Embeddings: Word meaning changes based on cotnext
    - BERT (By Google)
    - GPT (By OpenAI)

##### Applications of Word Embeddings
Word embeddings are used in
- Sentiment analysis
- Machine Translation
- Chatbots
- Search Engines
- Text Classification
- Named Entity Recognition

##### Advantages
- Capture semantic meaning
- Reduce dimensionality
- Improve model performance
- Enable transfer learning

##### Limitations
- Static embeddings ignore context
- Require large datasets
- Can inherit bias from training data

##### Example 1: Numpy

````python
import numpy as np

# Define a small vocabulary
word_embeddings = {
    "king": np.array([0.8, 0.2, 0.6]),
    "queen": np.array([0.75, 0.25, 0.65]),
    "man": np.array([0.9, 0.1, 0.4]),
    "woman": np.array([0.85, 0.15, 0.45])
}

# Example: similarity using cosine similarity
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

print("Similarity (king, queen):", cosine_similarity(word_embeddings["king"], word_embeddings["queen"]))
print("Similarity (man, woman):", cosine_similarity(word_embeddings["man"], word_embeddings["woman"]))
````

Example 2: Scikit-learn (CountVectorizer / TF-IDF)

````python
from sklearn.feature_extraction.text import TfidfVectorizer

sentences = [
    "I love NLP",
    "NLP is amazing",
    "I enjoy learning AI"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)

print(X.toarray())
print(vectorizer.get_feature_names_out())
````

Example 3: Tensorflow does not support python 3.14.x version

````python
import torch
import torch.nn as nn

embedding = nn.Embedding(100, 10)

input = torch.tensor([1, 2, 3])
output = embedding(input)

print(output)
````

Example 4: Use Pretrained Embeddings (No Training Required)

````python
from torchtext.vocab import GloVe

glove = GloVe(name='6B', dim=50)

print(glove['king'])
print(glove['queen'])
````
